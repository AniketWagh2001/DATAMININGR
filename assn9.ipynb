```r
library(arules)
library(arulesViz)
```

```r
basket <- read.transactions("basket.dat",format="basket", sep=" ", rm.duplicates = TRUE)
```

```r
rules1 = apriori(basket,parameter = list(supp=0.01,conf=0.3))
rules2 = apriori(basket,parameter = list(supp=0.02,conf=0.4))
rules3 = apriori(basket,parameter = list(supp=0.03,conf=0.5))
```

```r
plot(rules1)
```

```r
plot(rules2)
```

```r
plot(rules3)
```

```r
inspect(rules3[15:30])
```

```r
plot(rules3,method="grouped")
```
